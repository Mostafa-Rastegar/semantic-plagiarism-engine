# Exploration Notebook

این notebook تنها برای تست تعاملی پارامترها است. کد اصلی پروژه در `src/plagiarism_engine/` قرار دارد.

```python
from plagiarism_engine.preprocessing import preprocess_document, jaccard
from plagiarism_engine.minhash import MinHasher, estimated_jaccard
from plagiarism_engine.lsh import LSHIndex
from plagiarism_engine.simhash import SimHasher, simhash_similarity
from plagiarism_engine.preprocessing import tokenize

doc_a = 'Machine learning is a subset of artificial intelligence.'
doc_b = 'AI includes machine learning as a subfield.'

sa = preprocess_document(doc_a, shingle_size=2)
sb = preprocess_document(doc_b, shingle_size=2)
print('Jaccard exact:', jaccard(sa, sb))

mh = MinHasher(num_perm=128)
print('Jaccard MinHash est:', estimated_jaccard(mh.signature(sa), mh.signature(sb)))

sh = SimHasher(ngram=1, use_idf=True).fit([tokenize(doc_a), tokenize(doc_b)])
print('SimHash sim:', simhash_similarity(sh.sign(tokenize(doc_a)), sh.sign(tokenize(doc_b))))
```
